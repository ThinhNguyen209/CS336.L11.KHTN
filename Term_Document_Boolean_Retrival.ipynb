## BƯỚC 1: XÂY DỰNG TỪ ĐIỂN

```

# Đọc tất cả file text qua đường dẫn
# Lấy 20 file để test
file_paths= glob.glob('/content/gdrive/My Drive/Truy_van/Dataset_Crawl/*.txt')[:21]
#print(file_paths)

# xây dừng TỪ ĐIỂN 
lst_contents = []
dictionary = set()

for file_path in file_paths:
    f = open(file_path, 'r', encoding='utf-8')
    str = f.read()
    # lst_contents.append(str)
    info = str.split('\n')
    # lấy phần nội dung của test
    words = info[-1].replace('"', '').replace('.', '').replace("'","").split()
    words = set(words)
    lst_contents.append(words)
    dictionary.update(words)
    #print(words)
    #print ('Noi dung file cua ban la:\n', str)

print('Tap cac tu: ', dictionary)
```

## BƯỚC 2. XÂY DỰNG BOOLEAN FILE

```
# Tạo dict() của mỗi từ 
# Lưu số thứ tự của file chứa từ đang duyệt 
bool_files = dict()
for k,word in enumerate(dictionary):
    bool_files[word] = list()
    for i,content in enumerate(lst_contents):
        if word in content:
            bool_files[word].append(1)
        else:
            bool_files[word].append(0)
            #print(word,"    ",content)
            
print(bool_files)
```

## BƯỚC 3. XỬ LÍ CÂU TRUY VẤN

```
# Xây dựng các hàm logic
def Or (a,b):
  result = []
  for i,j in zip(a,b):
    if i == 0 and j == 0:
      result.append(0)
    else:
      result.append(1)
  return result

def And (a,b):
  result = []
  for i,j in zip(a,b):
    if i == 1 and j == 1:
      result.append(1)
    else:
      result.append(0)
  return result

def Xor (a,b):
  result = []
  for i,j in zip(a,b):
    if i == j:
      result.append(0)
    else:
      result.append(1)
  return result

def Not (a,b):
  result = []
  for i,j in zip(a,b):
    if i == 1 and j == 0:
      result.append(1)
    else:
      result.append(0)
  return result
```

```
string = 'quả and thức not Bạch or nước'

words = []
query_words = []
#Tách câu truy vấn thành 2 tập 
# 1 tập chứa các từ truy vấn or not and xor
# 2 là các từ 
string = string.split(' ')
for i in string:
    if i == 'or':
      query_words.append('|')
    elif i == 'and':
      query_words.append('&')
    elif i == 'not':
      query_words.append('-')
    elif i == 'xor':
      query_words.append('^')
    else:
      words.append(i)
print(query_words)
print(words)
# Tìm tập hợp của các từ 'words'
for i, word in enumerate(words):
	if word in bool_files:
		words[i] = bool_files[word]
	else:
    #Khử toán tử đi kèm vs từ đằng sau
    # nếu từ đằng sau ko có trong 'bool_files'
		del query_words[i-1]

print(words)
```

## BƯỚC 4. TRUY VẤN

```
result = words[0]
for i,word in enumerate(query_words):
  if word == '|':
    result = Or(result,words[i+1])
  elif word == '&':
    result = And(result,words[i+1])
  elif word == '-':
    result = Not(result,words[i+1])
  elif word == '^':
    result = Xor(result,words[i+1])
print(result)
```

## BƯỚC 5. XUẤT FILE 

```

for i,file in enumerate(result):
	if file == 1:
		f = open(file_paths[i], 'r', encoding= 'utf-8')
		print('\n'+ f.read())
```
