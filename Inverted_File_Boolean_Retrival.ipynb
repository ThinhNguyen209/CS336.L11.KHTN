## BƯỚC 1: XÂY DỰNG TỪ ĐIỂN

```

# Đọc tất cả file text qua đường dẫn
# Lấy 20 file để test
file_paths= glob.glob('/content/gdrive/My Drive/Truy_van/Dataset_Crawl/*.txt')[:21]
#print(file_paths)

# xây dừng TỪ ĐIỂN 
lst_contents = []
dictionary = set()

for file_path in file_paths:
    f = open(file_path, 'r', encoding='utf-8')
    str = f.read()
    # lst_contents.append(str)
    info = str.split('\n')
    # lấy phần nội dung của test
    words = info[-1].replace('"', '').replace('.', '').replace("'","").split()
    words = set(words)
    lst_contents.append(words)
    dictionary.update(words)
    #print(words)
    #print ('Noi dung file cua ban la:\n', str)

print('Tap cac tu: ', dictionary)
```

## BƯỚC 2. XÂY DỰNG INVERTED FILE

```
# Tạo dict() của mỗi từ 
# Lưu số thứ tự của file chứa từ đang duyệt 
inv_files = dict()
for k,word in enumerate(dictionary):
    inv_files[word] = set()
    for i,content in enumerate(lst_contents):
        if word in content:
            inv_files[word].add(i)
            #print(word,"    ",content)
            
print(inv_files)
```

## BƯỚC 3. XỬ LÍ CÂU TRUY VẤN

```
string = 'quả and thức not Bạch or nước'

words = []
query_words = []
#Tách câu truy vấn thành 2 tập 
# 1 tập chứa các từ truy vấn or not and xor
# 2 là các từ 
string = string.split(' ')
for i in string:
    if i == 'or':
      query_words.append('|')
    elif i == 'and':
      query_words.append('&')
    elif i == 'not':
      query_words.append('-')
    elif i == 'xor':
      query_words.append('^')
    else:
      words.append(i)
print(query_words)
print(words)
# Tìm tập hợp của các từ 'words'
for i, word in enumerate(words):
	if word in inv_files:
		words[i] = '{}'.format(inv_files[word])
	else:
		words[i] = '{}'

print(words)
```

## BƯỚC 4. TRUY VẤN

```
# Tạo chuỗi tập hợp và các phép or and 
# Vd: {1,3,2} or {}
str = words + query_words
str[::2] = words
str[1::2] = query_words

# Kết tập hợp vs phép toán lại
# Dùng hàm eval để chuyển string thành câu lệnh
result=eval(' '.join(str))
print(result)
```

## BƯỚC 5. XUẤT FILE 

```

for i in result:
 	f = open(file_paths[i], 'r', encoding= 'utf-8')
 	print('\n'+ f.read())
```
